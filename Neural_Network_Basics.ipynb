### Activation Function Tanh

### Loss Function

```python
def mse(h, y):
    return np.square(h - y)
```

### Last Layer Delta

```python
def last_layer_delta(xL, sL, y, act_deriv):
    return 2 * (xL - y) * act_deriv(sL)
```

### Network initialization

```python
def init_network(node_list):
    
    network = []
    
    for i in range(len(node_list) - 1):
        
        rows = node_list[i] + 1
        cols = node_list[i+1]
        
        weights = np.random.randn(rows, cols)
        
        network.append(weights)
        
    return network
```

### Forward Propagation

```python
def backprop(y, network, s_list, x_list, activations_deriv):
    
    deltas = []
    
    L = len(network)
    
    # last layer delta
    delta_L = last_layer_delta(
        x_list[-1],
        s_list[-1],
        y,
        activations_deriv[-1]
    )
    
    deltas.append(delta_L)
    
    for l in range(L-2, -1, -1):
        
        W = network[l+1][1:, :]
        
        delta = (W @ deltas[0]) * activations_deriv[l](s_list[l])
        
        deltas.insert(0, delta)
    
    return deltas
```

### Weight update

```python
def update_weights(network, x_list, deltas):
    
    for i in range(len(network)):
        
        prev_x = np.concatenate(([1], x_list[i]))
        
        grad = np.outer(prev_x, deltas[i])
        
        network[i] -= lr * grad
        
    return network
```
